```python
count=1             #so here we will use count and store 1 there
while count <= 5:   #now stating that count <= 5, this will print hello until and unless the value of count is 5 or less than 5
  print("hello")
  count += 1        #this will add 1 in count every time until annd unless count is 6 and terminates the loop as soon as it reaches 6 (<=5)
print(count)        #last value of count
```

```python
#search for a number x in this tuple using loop
#[1,4,9,16,25,36,49,64,81,100]
nums= (1,4,9,16,25,36,49,64,81,100)

x=(int(input("enter the number to search")))
i=0
while i<len(nums):
  if (nums[i]==x):
    print("found at index", i)
  i+=1
```

```python
#program to find the sum of first n numbers(using while loop)
n=5
sum = 0
i=1
while i<=n:
  sum+=i
  i+=1
print("total sum = ",sum)
```

```python
#write a program to print factorial of first n numbers(using for loop)
n=5
fact=1
for i in range(1,n+1):
  fact*=i
print(fact)
```
